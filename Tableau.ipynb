1. What is Tableau? Explain its importance in Business Intelligence and how
it helps in data-driven decision-making.

  -> Tableau is a data visualization and business intelligence tool that helps users turn raw data into clear and interactive dashboards. Instead of analyzing data in complex spreadsheets, Tableau presents information visually, making patterns, trends, and issues easy to understand. It connects to many types of data sources and updates insights in real time, which allows businesses to monitor performance as it happens.

Its importance in business intelligence comes from its ability to simplify data exploration and support fast decision-making. Managers and teams can use Tableau to identify opportunities, detect problems early, and understand customer or market behavior. Overall, Tableau helps organizations become more data-driven and respond quickly to changes in the business environment.

2. Explain the role of the following Tableau components:

a) Data Pane

b) Worksheet

c) Dashboard

d) Story

  -> Here is a clear explanation of the role of each Tableau component:

---

### **a) Data Pane**

The Data Pane is where all the fields from your data source appear after you connect to it. It contains dimensions (categories like Country or Product) and measures (numeric values like Sales or Profit). You use the Data Pane to drag fields onto shelves and create visualizations. It is basically your main toolbox for building charts.

---

### **b) Worksheet**

A Worksheet is a single canvas where you create one visualization at a time, such as a bar chart, scatter plot, or map. Each worksheet contains shelves and cards (like Rows, Columns, Filters, and Marks) where you place fields from the Data Pane to design your view. Each worksheet focuses on one specific chart or analysis.

---

### **c) Dashboard**

A Dashboard is a collection of multiple worksheets arranged together on a single screen. It allows you to combine charts, filters, images, and text to tell a complete story in one view. Dashboards are interactive—users can click on a chart to filter other charts, making analysis more dynamic and user-friendly.

---

### **d) Story**

A Story is a sequence of sheets or dashboards that presents insights step-by-step, like slides in a presentation. It helps explain data logically by walking the viewer through different points or findings. Stories are used to highlight insights, demonstrate trends, and support decision-making in a narrative format.

---


3. What is the difference between Dimensions and Measures in Tableau?
Provide examples of each.

  -> Dimensions

Dimensions are categorical fields that describe what you are analyzing. They usually contain text or discrete values and are used to slice, group, or label data. Dimensions create headers, labels, and categories in charts.

Examples of Dimensions:

- Country

- Product Name

- Region

- Customer ID

- Order Date (treated as a category when broken into year/month)

Measures

Measures are numeric fields that represent quantitative values that can be measured or aggregated. Tableau performs calculations like SUM, AVG, MIN, MAX on measures. Measures form the axes of charts.

Examples of Measures:

- Sales

- Profit

- Quantity

- Discount

- Cost

4. Define and explain the purpose of Filters, Parameters, and Sets in
Tableau.

   -> Here is a clear and simple explanation of each Tableau concept:

---

## **Filters**

**Definition:**
Filters are used to *limit or control which data appears* in a worksheet, dashboard, or story.

**Purpose:**
Filters help you focus on specific data by removing unwanted values. For example, you can filter data by year, country, product, or profit range. They allow users to refine what they see without changing the actual dataset.

**Example:**
Filtering the dashboard to show only “USA” sales or only “Consumer” segment.

---

## **Parameters**

**Definition:**
Parameters are *user-controlled input values* that can replace a constant in calculations, filters, reference lines, or other functions.

**Purpose:**
Parameters provide flexibility and interactivity. Unlike filters, which only limit existing data fields, parameters allow users to input values that *change the logic* of the analysis.

They can be used to:

* Switch between measures (Sales or Profit)
* Set a custom profit threshold
* Choose a specific date or product
* Control calculations dynamically

**Example:**
A parameter that allows the user to choose between “Show Sales” or “Show Profit” in a chart.

---

## **Sets**

**Definition:**
Sets are *custom groups* of data based on conditions or manual selection.

**Purpose:**
Sets help divide data into groups for comparison and deeper analysis. You can create dynamic sets based on conditions (e.g., top 10 customers by sales) or static sets by manually selecting items.

Sets are useful for:

* Finding top or bottom performers
* Highlighting a specific group
* Comparing selected vs unselected data

**Example:**
A set that identifies the “Top 10 Products by Sales” and highlights them in a visualization.

---

```python
plt.figure(figsize=(14,8))

bars = plt.bar(country_sales.index, country_sales.values, color='skyblue')

# Identify max and min
max_val = country_sales.max()
min_val = country_sales.min()

max_index = country_sales.idxmax()
min_index = country_sales.idxmin()

# Highlight max in red and min in green
for i, v in enumerate(country_sales.values):
    if v == max_val:
        bars[i].set_color('red')
    if v == min_val:
        bars[i].set_color('green')

# Add data labels
for i, v in enumerate(country_sales.values):
    plt.text(i, v, f"{v:,.0f}", ha='center', va='bottom', fontsize=9)

plt.xticks(rotation=45, ha='right')
plt.title("Gross Sales by Country (Max & Min Highlighted)", fontsize=16)
plt.xlabel("Country")
plt.ylabel("Gross Sales")

plt.tight_layout()
plt.show()
```

8. Create a dashboard that includes:

● KPI tiles for Total Sales, Total Profit, and Total Units Sold

● A line chart for Profit trend over time

● Filters for Product and Country

```python
!pip install plotly ipywidgets
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import ipywidgets as widgets
from IPython.display import display
```

```python
np.random.seed(42)

dates = pd.date_range(start="2023-01-01", periods=24, freq="M")
products = ["Laptop", "Tablet", "Phone"]
countries = ["USA", "Germany", "India"]

data = []

for date in dates:
    for product in products:
        for country in countries:
            units = np.random.randint(20, 200)
            sales = units * np.random.randint(300, 1200)
            profit = sales * np.random.uniform(0.15, 0.35)

            data.append([date, product, country, units, sales, profit])

df = pd.DataFrame(
    data,
    columns=["Date", "Product", "Country", "Units Sold", "Sales", "Profit"]
)

df.head()
```

```python
product_filter = widgets.Dropdown(
    options=["All"] + sorted(df["Product"].unique().tolist()),
    value="All",
    description="Product:"
)

country_filter = widgets.Dropdown(
    options=["All"] + sorted(df["Country"].unique().tolist()),
    value="All",
    description="Country:"
)
```

```python
def update_dashboard(product, country):
    filtered_df = df.copy()

    if product != "All":
        filtered_df = filtered_df[filtered_df["Product"] == product]

    if country != "All":
        filtered_df = filtered_df[filtered_df["Country"] == country]

    # KPIs
    total_sales = filtered_df["Sales"].sum()
    total_profit = filtered_df["Profit"].sum()
    total_units = filtered_df["Units Sold"].sum()

    # Profit Trend
    trend = (
        filtered_df
        .groupby("Date")["Profit"]
        .sum()
        .reset_index()
    )

    # Create Figure
    fig = make_subplots(
        rows=2, cols=3,
        specs=[[{"type": "indicator"}, {"type": "indicator"}, {"type": "indicator"}],
               [{"colspan": 3, "type": "scatter"}, None, None]],
        subplot_titles=("Total Sales", "Total Profit", "Total Units Sold", "Profit Trend Over Time")
    )

    fig.add_trace(go.Indicator(
        mode="number",
        value=total_sales,
        number={'prefix': "$", 'valueformat': ',.0f'}
    ), row=1, col=1)

    fig.add_trace(go.Indicator(
        mode="number",
        value=total_profit,
        number={'prefix': "$", 'valueformat': ',.0f'}
    ), row=1, col=2)

    fig.add_trace(go.Indicator(
        mode="number",
        value=total_units,
        number={'valueformat': ',.0f'}
    ), row=1, col=3)

    fig.add_trace(go.Scatter(
        x=trend["Date"],
        y=trend["Profit"],
        mode="lines+markers",
        name="Profit"
    ), row=2, col=1)

    fig.update_layout(height=600, showlegend=False)
    fig.show()
```

```python
widgets.interactive(
    update_dashboard,
    product=product_filter,
    country=country_filter
)
```

9. Your goal is to identify products that generate low profit despite high sales
volume.

● Use scatter plot or highlight table to identify such products.

● Add filters for Country and Segment.

● Write two business insights based on your chart.

```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display
```

```python
df = pd.read_excel("Global_Sales_dataset.xlsx")
df.head()
```

```python
product_perf = (
    df.groupby(["Product", "Country", "Segment"], as_index=False)
    .agg({
        "Units Sold": "sum",
        "Gross Sales": "sum",
        "Profit": "sum"
    })
)
```

```python
import plotly.express as px

def update_scatter(country, segment):
    filtered = product_perf.copy()

    if country != "All":
        filtered = filtered[filtered["Country"] == country]

    if segment != "All":
        filtered = filtered[filtered["Segment"] == segment]

    fig = px.scatter(
        filtered,
        x="Units Sold",
        y="Profit",
        size="Gross Sales",
        color="Product",
        hover_data=["Gross Sales"],
        title="High Sales Volume vs Low Profit Products"
    )

    fig.update_layout(height=600)
    fig.show()
```

```python
import ipywidgets as widgets
from IPython.display import display

country_filter = widgets.Dropdown(
    options=["All"] + sorted(df["Country"].unique()),
    description="Country:"
)

segment_filter = widgets.Dropdown(
    options=["All"] + sorted(df["Segment"].unique()),
    description="Segment:"
)

widgets.interactive(update_scatter, country=country_filter, segment=segment_filter)
display(country_filter, segment_filter)
```
